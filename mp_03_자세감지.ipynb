```python
def download_model(model_path, model_url):
    # 기존 모델은 다운로드 생략
    if model_path.exists():
        return

    # 모델이 없을 때만 다운로드
    print(f"모델 다운로드 중: {model_path}")
    urllib.request.urlretrieve(model_url, model_path)


# 첫 실행 시 자세 모델 다운로드
download_model(POSE_LANDMARKER_MODEL, POSE_LANDMARKER_MODEL_URL)
```
