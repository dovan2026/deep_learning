### 1. 기본 얼굴 검출

- Haar Cascade
    - 기계 학습 기반의 객체 탐지 알고리즘으로, 이미지 또는 비디오에서 얼굴, 눈, 차량 번호판 등의 특정 패턴을 감지하는 데 사용
    - 이 알고리즘은 Haar-like 특징(Haar-like Features)과 Cascade 분류기(Cascade Classifier) 를 조합하여 빠르고 효율적인 검출을 수행
    - 딥러닝 모델보다 가볍지만 조명, 얼굴 각도, 가림에 민감할 수 있음

- 다른 Cascade 파일 사용 가능

| Cascade 파일명                      | 검출 대상       |
|-------------------------------------|---------------|
| `haarcascade_frontalface_default.xml` | 정면 얼굴 검출 |
| `haarcascade_eye.xml`                | 눈 검출       |
| `haarcascade_smile.xml`              | 웃는 얼굴 검출 |
| `haarcascade_fullbody.xml`           | 전신 검출     |
| `haarcascade_upperbody.xml`          | 상반신 검출   |


```python
import cv2
```

```python
# OpenCV에 포함된 haarcascade 폴더 경로 확인
# 이 경로 뒤에 사용할 CML 파일명을 붙여서 전달.
print(cv2.data.haarcascades)
```

```python
cascade_path = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'

# Haar Cascade 로드
# Cascade 분류기를 로드
face_cascade = cv2.CascadeClassifier(cascade_path)
```

```python
# 웹캠 열기
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("웹캠이 연결이 안됩니다")

while True:

    # 프레임 읽기
    ret, frame = cap.read()

    if not ret:
        break

    # 그레이스케일 - 흑백 변환
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # 얼굴 검출
    # detectMultiScale : 이미지 크기를 여러 단계로 줄여가며 얼굴 후보를 찾음
    faces = face_cascade.detectMultiScale(
        gray_frame,
        scaleFactor=1.1,                # 이미지 축소 비율. 1.1이면 10%씩 줄이며 탐색.
        minNeighbors=5,                 # 후보 주변에 이웃 검출이 얼마나 있어야 인정할지 정함. 클수록 보수적.
        minSize=(300,300))              # 이 크기보다 작은 얼굴 후보는 무시
    
    # 검출된 각 얼굴 영역에 사각형 그리기
    for x, y, w, h in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)

    # 출력
    cv2.imshow("frame", frame)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

### 2. 그냥 합성해보기

```python
# 넣을 이모티콘 이미지 로드
emoji_img = cv2.imread('data/person_ma.jpg')
cv2.imshow('emoji_img', emoji_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 웹캠 열기
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("웹캠이 연결이 안됩니다")

while True:

    # 프레임 읽기
    ret, frame = cap.read()

    if not ret:
        break

    # 그레이스케일 - 흑백 변환
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # 얼굴 검출
    faces = face_cascade.detectMultiScale(gray_frame,
                                          scaleFactor=1.1,
                                          minNeighbors=5,
                                          minSize=(30,30))

    for x, y, w, h in faces:
        # 검출된 얼굴 크기에 맞게 합성 이미지를 조정
        emoji_img_resized = cv2.resize(emoji_img, (w, h))

        # 얼굴 영역을 합성 이미지로 그대로 교체
        frame[y:y + h, x:x + w] = emoji_img_resized

    # 출력
    cv2.imshow("frame", frame)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

### 3. 자연스럽게 다양한 효과를 적용해서 합성해보기

```python
# 넣을 이모티콘 이미지 로드
emoji_img = cv2.imread('data/cat_face.png', cv2.IMREAD_UNCHANGED)
print(emoji_img.shape)
cv2.imshow('emoji_img', emoji_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np

# 4번쨰 채널(투명도)만 가져옴
# 255로 나눠서 0.0 ~ 1.0 범위로 정규화
emoji_alpha = emoji_img[:, :, 3] / 255.0

# 일부 영역의 알파값 확인
emoji_alpha[100:200, 200:300]
```

```python
emoji_bgr = emoji_img[:, :, :3] # brg 채널값
emoji_bgr[100:200, 200:300]
```

```python
# 웹캠 열기
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("웹캠이 연결이 안됩니다")

while True:

    # 프레임 읽기
    ret, frame = cap.read()

    if not ret:
        break

    # 그레이스케일 - 흑백 변환
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # 얼굴 검출
    faces = face_cascade.detectMultiScale(gray_frame,
                                          scaleFactor=1.1,
                                          minNeighbors=5,
                                          minSize=(30,30))

    for i in faces:
        x, y, w, h = i

        # 넣을 이미지 크기 조정
        emoji_img_resized = cv2.resize(emoji_bgr, (w, h))

        # 알파 채널도 이미지 크기 조정
        emoji_alpha_resized = cv2.resize(emoji_alpha, (w, h))

        # 얼굴 부위 추출
        face_part = frame[y:y+h, x:x+w]

        # 배경(얼굴 부위) 반대 마스크
        background_alpha = 1.0 - emoji_alpha_resized

        # 마스크 이미지의 알파 채널과 RGB 채널을 곱해서 마스크 이미지에서 알파 처리된 부분을 추출
        alpha_face = emoji_alpha_resized[..., np.newaxis] * emoji_img_resized

        # 원본 이미지의 해당 영역과 배경 알파 채널을 곱해서 배경 추출
        background_face = background_alpha[..., np.newaxis] * face_part

        # 최종 추출된 두 부분을 합쳐서 얼굴 부분을 생성
        face_part = alpha_face + background_face

        frame[y:y+h, x:x+w] = face_part
    
    cv2.imshow("frame", frame)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
