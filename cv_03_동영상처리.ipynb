### 1. 영상 불러오기 및 정보 확인

```python
import cv2
```

```python
cap = cv2.VideoCapture('data/cat_video.mp4')
```

```python
print("영상의 가로 크기", int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)))
print("영상의 세로 크기", int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
print("영상의 프레임수", int(cap.get(cv2.CAP_PROP_FRAME_COUNT)))
print("영상의 프레임 속도", int(cap.get(cv2.CAP_PROP_FPS)))
```

### 2. 영상 출력해보기

```python
while True:
    # ret : 프레임 읽기 성공 여부
    # frame : 실제 프레임 이미지 데이터, NumPy 배열.
    # cap.read() : 한 번 호출할 때마다 다음 프레임 한 장을 반환.
    ret, frame = cap.read()
    
    # 더 이상 읽을 프레임이 없거나 오류가 나면 반복을 중단
    if not ret:
        break
    
    # 읽은 프레임을 OpenCV 창에 표시
    cv2.imshow('frame', frame)

    # q 입력하면 종료
    if cv2.waitKey(1) == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()
```

### 3. 웹캠으로 카메라 정보를 출력해보기

```python
# 0번 장치를 기본 웹캠으로 엽니다.
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break
    # 프레임을 표시
    else:
        # 웹캠은 거울처럼 보이는 것이 자연스러우므로 좌우 반전.
        frame = cv2.flip(frame,1)

        # 얼굴이 있을 것으로 가정한 임의의 영역 잘라냄
        face = frame[100:300, 200:400]

        # 작은 크기로 줄이면 세부 정보가 사라짐
        face_resized = cv2.resize(face, (10, 10), interpolation=cv2.INTER_AREA)

        # 다시 원래 영역 크기로 확대하면 픽셀이 크게 보여 모자이크처럼 보임.
        face_resized = cv2.resize(face_resized, (200, 200), interpolation=cv2.INTER_AREA)

        # 처리한 영역을 원봄 프레임의 같은 위치에 덮어씁니다.
        frame[100:300, 200:400] = face_resized

        cv2.imshow("frame", frame)
        cv2.imshow('face_resized', face_resized)
        
        if cv2.waitKey(1) == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```

### 4. 영상 저장하기

```python
cap = cv2.VideoCapture('data/cat_video.mp4')
```

```python
# 원본 영상과 같은 크기/FPS로 저장하기 위해 속성을 읽음
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
```

- 비디오 코덱 설정

| 코덱   | 설명                                 |
|--------|--------------------------------------|
| `'XVID'` | Xvid (압축률 좋음, AVI 파일)        |
| `'MP4V'` | MPEG-4 (MP4 파일)                 |
| `'X264'` | H.264 (고품질 압축)                |
| `'MJPG'` | Motion JPEG (무손실, 용량 큼)      |


```python
# 비디오 저장을 위한 videowriter 객체 설정
# fourcc = cv2.VideoWriter_fourcc(*'MP4V')
fourcc = cv2.VideoWriter_fourcc(*'X264')
out = cv2.VideoWriter('gray_cat.mp4', fourcc, fps, (frame_width, frame_height), False)
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break
    # 프레임을 표시
    else:
        # 흑백으로 바꾸기
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        out.write(frame)

        cv2.imshow("frame", frame)
        if cv2.waitKey(1) == ord('q'):
            break
cap.release()
out.release()
cv2.destroyAllWindows()
```
